```python
def read_orbit(snap):
    orbit = np.loadtxt(snap)
    t = orbit[:,0]
    x = orbit[:,1]
    y = orbit[:,2]
    z = orbit[:,3]
    return t, x, y, z
```

```python
static = read_orbit('./test_obit_LMC6')

t_st = static[0]
x_st = static[1]
y_st = static[2]
z_st = static[3]
```

```python
tevol = read_orbit('./test_obit_LMC6_tevol')

t_t = tevol[0]
x_t = tevol[1]
y_t = tevol[2]
z_t = tevol[3]
```
